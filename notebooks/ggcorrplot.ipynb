## ggcorrplot 

In this notebook, we’ll describe the python package **ggcorrplot** for displaying easily a correlation matrix using ‘plotnine’.

### mtcars dataset

the [mtcars](https://plotnine.org/reference/mtcars.html#plotnine.data.mtcars) data set will be used in the following python code. The function **cor_pmat** [in **ggcorrplot**] computes a matrix of correlation p-values

```python
#disable warnings
from warnings import simplefilter, filterwarnings
simplefilter(action='ignore', category=FutureWarning)
filterwarnings("ignore")
```

```python
#load mtcars dataset form plotnine
from plotnine.data import mtcars
print(mtcars)
```

We set name as rownames

```python
# Set name as index
mtcars = mtcars.set_index("name")
print(mtcars)
```

### Compute a correlation matrix

```python
# Compute a correlation matrix
corr = mtcars.corr()
print(corr.round(4))
```

### Correlation matrix p-value

```python
# Compute a matrix of correlation p-values
from ggcorrplot import cor_pmat
p_mat = cor_pmat(mtcars)
print(p_mat.round(4))
```

### Correlation matrix visualization

```python
from ggcorrplot import ggcorrplot
```

#### method = "squared" (default)

```python
# method = "square" (default)
p = ggcorrplot(corr)
p
```

#### method = "circle"

```python
# method = "circle"
p = ggcorrplot(corr,method = "circle")
p
```

### Reordering the correlation matrix

```python
# using hierarchical clustering
p = ggcorrplot(corr, 
               hc_order = True, 
               outline_color = "white")
p
```

### Types of correlogram layout

#### Get the lower triangle

```python
# Get the lower triangle
p = ggcorrplot(corr, 
               hc_order = True, 
               type = "lower",
               outline_color = "white")
p
```

#### Get the upper triangle

```python
# Get the upper triangle
p = ggcorrplot(corr, 
               hc_order = True, 
               type = "upper",
               outline_color = "white")
p
```

#### Change colors and theme

```python
# Argument colors
from plotnine import theme_gray
p = ggcorrplot(corr, 
               hc_order = True, 
               type = "lower",
               outline_color = "white",
               ggtheme = theme_gray(),
               colors = ("#6D9EC1", "white", "#E46726"))
p
```

### Add correlation coefficients

```python
# argument label = True
p = ggcorrplot(corr, 
               hc_order = True, 
               type = "lower", 
               label = True)
p
```

### Add correlation significance level

```python
# Argument p_mat
# Barring the no significant coefficient
p = ggcorrplot(corr, 
               hc_order = True,
               type = "lower", 
               p_mat = p_mat)
p
```

#### Leave blank on no significant coefficient

```python
# Leave blank on no significant coefficient
p = ggcorrplot(corr, 
               p_mat = p_mat, 
               hc_order = True,
               type = "lower", 
               insig = "blank")
p
```

### Using original data

```python
#usinf original dataset
p = ggcorrplot(mtcars,
               matrix_type="completed")
p
```
